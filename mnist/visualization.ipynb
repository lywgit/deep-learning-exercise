# Visualizing high dimensional data

* PCA
* t-SNE

```python
import numpy as np
from sklearn.manifold import TSNE
from sklearn.datasets import load_iris, load_digits
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

from helper import read_mnist_data
%matplotlib inline
```

```python
DATA_PATH = '/Users/lywang/Data/mnist'
image_train, label_train, image_test, label_test = read_mnist_data(DATA_PATH)
```

```python
# PCA
X_pca = PCA(n_components=2).fit_transform(image_test)
```

```python
plt.scatter(X_pca[:,0], X_pca[:,1], c=np.squeeze(label_test));
```

```python
h = plt.scatter(X_pca[:,0], X_pca[:,1], c=np.squeeze(label_test))
plt.legend()
```

```python
# first rediuce to 10 dim using PCA to save computation time (still take a while)
XX = PCA(n_components=10).fit_transform(image_test) 
X_tsne = TSNE(n_components=2,random_state=33).fit_transform(XX)
```

```python
plt.scatter(X_tsne[:,0], X_tsne[:,1], c=np.squeeze(label_test))
```
