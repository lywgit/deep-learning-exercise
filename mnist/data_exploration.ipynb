```python
import matplotlib as matplotlib 
import matplotlib.pyplot as plt

from helper import read_mnist_data
%matplotlib inline
```

```python
DATA_PATH = '/Users/lywang/Data/mnist'
image_train, label_train, image_test, label_test = read_mnist_data(DATA_PATH)
```

```python
def show_image(i, image, label=None):
    print('image number {}:'.format(i) )
    if label is not None:
        print('label = {}'.format(label[i][0]))
    plt.figure()
    plt.imshow(image[i,:].reshape(28, 28), cmap = matplotlib.cm.gray);
    if label is not None:
        plt.title('True label:'+str(label[i][0]), fontsize=15)
    plt.show()

for i in [0, 1000, 2000, 5000]:
    show_image(i, image_train, label_train)
    show_image(i, image_test, label_test) 
```
